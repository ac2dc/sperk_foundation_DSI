## Technical Stack  : Scikit Learn, Numpy Array, Scipy, Pandas, Matplotlib

## Importing the Libraries required for analysis

```python
from sklearn import datasets
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.cluster import KMeans
```

## Load the Data

```python
iris = datasets.load_iris()
```

## Define target and predictors

```python
X = iris.data[:, :2] 
y = iris.target
```

## Observing Data using scatter plot

```python
plt.scatter(X[:,0], X[:,1], c=y, cmap='gist_rainbow')
plt.xlabel('Spea1 Length', fontsize=8)
plt.ylabel('Sepal Width', fontsize=8)
```

## Implementing K-means clustering

```python
sum_of_sqr_distance = []

for i in range(1, 11):
    kmeans = KMeans(n_clusters = i, init = 'k-means++', max_iter = 300, n_init = 10, random_state = 0)
    kmeans.fit(x)
    sum_of_sqr_distance.append(kmeans.inertia_)


kmeans = KMeans(n_clusters = 3, init = 'k-means++', max_iter = 300, n_init = 10, random_state = 0)
y_kmeans = kmeans.fit_predict(x)
```

## Using elbow method to determine optimal number of clusters

```python
plt.plot(range(1, 11), sum_of_sqr_distance)
plt.title('The elbow method')
plt.xlabel('Number of clusters')
plt.ylabel('sum of square distance')
plt.show()
```

## Visualizing the Clusters and ploting Centroids
